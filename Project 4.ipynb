```python


# Select only numeric columns
numeric_columns = df.select_dtypes(include=['number']).columns
numeric_df = df[numeric_columns]

# Get the correlation matrix
correlation_matrix = numeric_df.corr()

# Set the threshold for correlation
threshold = 0.8

# Find the pairs of variables with correlation greater than the threshold
highly_correlated_pairs = (correlation_matrix.abs() > threshold) & (correlation_matrix < 1)


highly_correlated_variables = set()

for col in correlation_matrix.columns:
    correlated_cols = correlation_matrix.index[highly_correlated_pairs[col]].tolist()
    if correlated_cols:
        correlated_vars = [col] + correlated_cols
        highly_correlated_variables.add(tuple(sorted(correlated_vars)))

# Display the sets of highly correlated variables
for var_set in highly_correlated_variables:
    print(f"Highly correlated variables: {var_set}")
```
